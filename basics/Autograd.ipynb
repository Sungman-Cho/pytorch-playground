```
import torch
import torch.nn as nn
```

# requires_grad 조작하기

```
x = torch.ones(2, 2, requires_grad=True)
print(x)
x.requires_grad_(False)
print(x)
```

# .detach() 조작하기

```
x = torch.ones(2, 2, requires_grad=True)
print(x.requires_grad)
y = x.detach()
print(y.requires_grad)
print(x.requires_grad)
```

# with torch.no_grad() 활용하기

```
x = torch.ones(2, 2, requires_grad=True)
print(x.requires_grad)
y = x.detach()
print(y.requires_grad)
y.requires_grad_(True)
print(y.requires_grad)

with torch.no_grad():
    z = x+y
    print(z.requires_grad)
```

# model.eval() 확인해보기

```
drop = nn.Dropout()
x = torch.ones(1, 10)

# Train
drop.train()
print(drop(x))

# Eval
drop.eval()
print(drop(x))
```
